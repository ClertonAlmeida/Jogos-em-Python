# Jogo da Forca

## Criando a Função

```python
import random
  
def jogar():
    
    menssagem() #chamando a função da mensagem inicial
    palavra_secreta = carrega_palavra() #chamando a função para carregar a palavra secreta
    letra_certa = inicializa_letra_certa(palavra_secreta) #chamando a função para inicializar a letra correta
    print(letra_certa)
    
    enforcou = False
    acertou = False
    erros = 0
    # Loop para verificação - enquanto for verdadeiro  os continuar jogando
    while (not enforcou and not acertou):
        
        #pedindo o usuario para chutar a letra
        chute = input("Digite a letra! ").upper().strip()
        
        if (chute in palavra_secreta):
            marcando_chute(chute, letra_certa, palavra_secreta) #chamando a função para marca que a letra esta correta
        else:
            erros += 1
            desenha_forca(erros) #chamando a função forca
        #Encerrando o loop    
        enforcou = erros == 7
        acertou = "_" not in letra_certa
        print(letra_certa)
   
    #Informando se ganhou ou perdeu
    if (acertou):
        imprime_vencedor() #chamando a função vencedor
    else:    
        imprime_perdedor(palavra_secreta) #chamando a função perdedor
   
    
def desenha_forca(erros):
    print("  _______     ")
    print(" |/      |    ")

    if(erros == 1):
        print(" |      (_)   ")
        print(" |            ")
        print(" |            ")
        print(" |            ")

    if(erros == 2):
        print(" |      (_)   ")
        print(" |      \     ")
        print(" |            ")
        print(" |            ")

    if(erros == 3):
        print(" |      (_)   ")
        print(" |      \|    ")
        print(" |            ")
        print(" |            ")

    if(erros == 4):
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |            ")
        print(" |            ")

    if(erros == 5):
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |       |    ")
        print(" |            ")

    if(erros == 6):
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |       |    ")
        print(" |      /     ")

    if (erros == 7):
        print(" |      (_)   ")
        print(" |      \|/   ")
        print(" |       |    ")
        print(" |      / \   ")

    print(" |            ")
    print("_|___         ")
    print()    

def imprime_vencedor():
    print("Parabéns, você ganhou!")
    print("       ___________      ")
    print("      '._==_==_=_.'     ")
    print("      .-\\:      /-.    ")
    print("     | (|:.     |) |    ")
    print("      '-|:.     |-'     ")
    print("        \\::.    /      ")
    print("         '::. .'        ")
    print("           ) (          ")
    print("         _.' '._        ")
    print("        '-------'       ")
    
def imprime_perdedor(palavra_secreta):
    print("Puxa, você foi enforcado!")
    print("A palavra era {}".format(palavra_secreta))
    print("    _______________         ")
    print("   /               \       ")
    print("  /                 \      ")
    print("//                   \/\  ")
    print("\|   XXXX     XXXX   | /   ")
    print(" |   XXXX     XXXX   |/     ")
    print(" |   XXX       XXX   |      ")
    print(" |                   |      ")
    print(" \__      XXX      __/     ")
    print("   |\     XXX     /|       ")
    print("   | |           | |        ")
    print("   | I I I I I I I |        ")
    print("   |  I I I I I I  |        ")
    print("   \_             _/       ")
    print("     \_         _/         ")
    print("       \_______/           ")
    
def menssagem():  
    print("*********************************")
    print("***Bem vindo ao jogo da Forca!***")
    print("*********************************")

def carrega_palavra():
      
    abrir_arquivo = open("palavras.txt", "r")
    lista_palavra = []
    #armazenando as palavras em uma lista
    for linha in abrir_arquivo:
        linha = linha.strip()
        lista_palavra.append(linha)
    abrir_arquivo.close() 
    #escolhendo a palavra aleatoriamente e informando a posição para armazenar ela na variavel
    num = random.randrange(0, len(lista_palavra))
    palavra_secreta = lista_palavra[num].upper()
            
    return palavra_secreta
   
def inicializa_letra_certa(palavra):
    return ["_" for letra in palavra]

def marcando_chute(chute, letra_certa, palavra_secreta):
     # Index para verificar a posição da letra dentro do loop que esta verificando se tem a letra dentro da palavra secreta
            index = 0
            for letra in palavra_secreta:
                if (chute == letra):
                    letra_certa [index] = chute #preenchendo nossa lista com as letras corretas
                index += 1
    
# Chamando a função main
if(__name__ == "__main__"):
    jogar()
```
